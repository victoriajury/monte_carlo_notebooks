```python
import numpy as np

import matplotlib.animation as animation
import matplotlib.pyplot as plt
```

```python
# initialize variables
n_simulations = 2000
n_points_circle = 0
n_points_square = 0

# create lists to store x and y values
l_xs = []
l_ys = []
```

```python
# loop n_simulations times
for _ in range(n_simulations):
    
    # x is randomly drawn from a continuous uniform distritbuion
    x = np.random.uniform(-1, 1)
    # store x in the list
    l_xs.append(x)
    
    # y is randomly drawn from a continuous uniform distribution
    y = np.random.uniform(-1, 1)
    # store y in the list
    l_ys.append(y)
```

```python
circle_x = []
circle_y = []

# loop n_simulations times
for i in range(n_simulations):

    # calculate the distance between the point and the origin
    dist_from_origin = l_xs[i] ** 2 + l_ys[i] ** 2
    
    # if the distance is smaller than or equal to 1, the point is in the circle
    if dist_from_origin <= 1:
        n_points_circle += 1
        circle_x.append(l_xs[i])
        circle_y.append(l_ys[i])

    # by definition of the uniform distribution, the point is in the square
    n_points_square += 1
```

```python
# estimate the value of pi
pi = 4 * n_points_circle / n_points_square
print(pi)
```

```python
# animate plot

def animate(i, cx=[], cy=[], sx=[], sy=[]):
    plt.cla()
    if l_xs[i] in circle_x:
        cx.append(l_xs[i])
    else:
        sx.append(l_xs[i])
    if l_ys[i] in circle_y:
        cy.append(l_ys[i])
    else:
        sy.append(l_ys[i])
    
    plt.scatter(cx, cy, color="red")
    plt.scatter(sx, sy, color="blue")
    

fig = plt.figure()
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
ani = animation.FuncAnimation(fig, animate, interval=1, save_count=n_simulations)
ani.save("circle_plot.gif")
plt.show()
```
